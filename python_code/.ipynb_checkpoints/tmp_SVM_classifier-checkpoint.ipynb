# TRAINING A SUPPORT VECTOR MACHINE



In order to improve our simple, inharmonicity-based classifier, we aim to build a more complete system that uses several other __audio features__ to perform the classification. Inspired by Jakob Abesser, _Automatic String Detection for Bass Guitar and Electric Guitar_, we chose several other features, presented in a previous notebook. 

These time- and frequency-domain characteristics might be used to distinguish between different strings/fingerings. Indeed, it is reasonable to think that different strings (and all related properties, e.g material, thickness, etc...) will attenuate more or less some partials, produce a more or less inharmonic spectrum, decay more or less rapidly, and so on. 

Our fingering estimation problem boils down to estimating the string number, ranging from 1 to 6. Therefore, we can formulate our problem as a __multi-class classification problem__. The 6 different classes (or labels) then correspond to the different strings. 

One possibility is to train a __Support Vector Machine (SVM)__, which is known to perform well for several classification tasks. The support vector machine tries to find the best separating hyperplanes (with a maximal margin) in the __feature space__, in order to separate the 6 strings.

Moreover, as we saw, for a given fundamental frequency $f_0$, only a few positions on the fretboard are possible (assuming a fixed, standard tuning). Using the particular geometry of the guitar, we can therefore safely discard any class/string that is not "playable". We will make use of this knowledge when analyzing the classes probability in the sequel.

We will first process all our recordings, extract the various audio features, and stack these into a matrix $X$. The latter will then be divided into a training matrix and a test matrix. The former will be used to train our classifier, for which several parameters need to be tuned (see later). We will then proceed along the same lines with the test recordings and test our classifier on our entire test set.

## I - EXTRACTING ALL THE FEATURES

From our entire dataset (consisting of 780 recordings), we first need to extracture all the desired audio features, as discussed above. We will construct a fairly large matrix $X$, where each row represents a sample, and each column a feature. We will take a look at the variation of the features, to grasp which one might be better/more discriminative for later use in our classifier.

### I.I - BUILDING THE MATRIX $X$

```python
# initializations
X = np.zeros([6 * 13 * num_rec_per_pos, num_features])
true_labels = np.hstack((np.ones(num_rec_per_pos*13), 2 * np.ones(num_rec_per_pos*13), 3 * np.ones(num_rec_per_pos*13), \
                   4 * np.ones(num_rec_per_pos*13), 5 * np.ones(num_rec_per_pos*13), 6 * np.ones(num_rec_per_pos*13)))

print("\nExtracting the audio features...")

idx = 0

# iterate over all possible fingerings
for n in range(6):
    
    print("\nString n°{str_num}".format(str_num=n+1))
    
    for m in range(13):
        
        print("\n\tFret n°{fret_num}".format(fret_num=m))
        
        # initializations for the current fingering
        file_folder = 'string_' + str(n + 1) + '/string_' + str(n + 1) + '_fret_' + str(m) + '/'
        file_name_prefix = 'string_' + str(n + 1) + '_fret_' + str(m) + '_take_'
        
        for rec_num in range(num_rec_per_pos):
            
            # read the recording from the data folder
            file_name = file_name_prefix + str(int(rec_num + 1)) + '.wav'            
            fs_recording, current_recording = wavfile.read(PATH_RECORDINGS + file_folder + file_name)
            y = current_recording / max(current_recording)
            
            # extract the first 1.5 seconds of the recording
            y = y[:int(1.5*fs_recording)]
            
            # TMP: PARAMETERS (TO PUT SOMEWHERE ELSE)
            tot_num_samp = y.shape[0]
            num_samp_per_frame = int(fs_recording * frame_duration)
            window = np.hamming(num_samp_per_frame)
            hop_size = int(num_samp_per_frame / 2)
            num_frames = int((tot_num_samp - num_samp_per_frame) / hop_size)
            
            # estimate the fundamental frequency
            estimated_fundamental = estimate_fundamental(np.abs(np.fft.rfft(np.hamming(y.shape[0])*y, NFFT)), fs_recording, \
                                                 NFFT, tol, max_num_peaks)
            # TMP
            if(n == 0 and m == 0):
                estimated_fundamental = 329.6
            #print("\n\t\tEstimated fundamental [Hz]: {est_fund}".format(est_fund=estimated_fundamental))
            
            # track the partials in successive time frames with our modified Goertzel
            estimated_partials, partials_amplitudes = partials_tracking(y, fs_recording, frame_duration, \
                                                                        estimated_fundamental, max_mode, delta_f, 'goertzel')
            
            # extract all the audio features from the current recording
            curr_feature_vec = extract_audio_features(y, fs_recording, estimated_partials, partials_amplitudes, \
                                                      max_mode, num_frames, 10)
            
            # fill in the matrix X with the current example feature vector
            X[idx, :] = curr_feature_vec
            idx = idx + 1
            
        print("\t\tDone!")
```

### I.II - FEATURES VISUALIZATION

We take a look at the variation of each feature for the different strings of the guitar, denoted by their pitch in the following (E, B, G, D, A, E, from the first to the last respectively). If some features are not that discriminative, we might discard them in a later dimensionality reduction step (see below).

```python
plt.figure(figsize=(20, 8))

# 1st most discriminative feature
feature_1 = -1
tmp = np.zeros([13 * num_rec_per_pos, 6])
for n in range(1, 7):
    tmp[:, n - 1] = X[(n - 1) * 13 * num_rec_per_pos:n * 13 * num_rec_per_pos, feature_1]

ax1 = plt.subplot2grid((2, 2), (0, 0), rowspan=2, colspan=1)
ax1.boxplot(tmp, 0, '', labels=['E', 'B', 'G', 'D', 'A', 'E'])
plt.title('Inharmonicity coefficient')

# 2nd most discriminative feature
feature_2 = 0
for n in range(1, 7):
    tmp[:, n - 1] = X[(n - 1) * 13 * num_rec_per_pos:n * 13 * num_rec_per_pos, feature_2]
    
ax2 = plt.subplot2grid((2, 2), (0, 1), rowspan=2, colspan=1)
ax2.boxplot(tmp, 0, '', labels=['E', 'B', 'G', 'D', 'A', 'E'])
plt.title('Global Decay Time')

plt.tight_layout()
plt.show()
```

The inharmonicity coefficient seems to be the most distinctive/discriminative feature (which is not that surprising from our previous studies on inharmonicity and our inharmonicity-based classifier). The global decay time might also be used to distinguish different strings/fingerings. Even though it is less clear for the partials' amplitudes, it seems that there is still some variation between the strings.

### I.III - DIMENSIONALITY REDUCTION USING SVD

It might be beneficial (both computationally and for the actual results) to reduce the dimension of our matrix X, by combining some features in a way as to keep the most relevant information. This can be done using a "truncated" Singular Value Decomposition (SVD).

```python
# singular value decomposition
U, s, V = np.linalg.svd(X)

# visualize the variance of each dimension
plt.figure()
plt.stem(s)
plt.xlabel('Dimension "number"')
plt.ylabel('"Variance/Information"')
plt.title('Variance of each dimension/feature')
plt.tight_layout()
plt.show()
```

It seems that the first 3-5 dimensions contain most of the variance/information.

```python
# compute a low-rank approximation of our matrix X
K = 3
low_rank_X = X @ V[:, :K]
```

## II - TRAIN-TEST SPLIT

We now split the data into a training set (a matrix $X_{train}$) and a test set (a matrix $X_{test}$). For now, we have chosen a 50-50 split (50% for both training and testing).

```python
# Each entry is an array with the recording indices/numbers
# used for training and testing, respectively
train_indices = np.zeros([6, 13, 5])
test_indices = np.zeros([6, 13, 5])

X_train = np.zeros([6 * 13 * 5, num_features])
X_train_lr = np.zeros([6 * 13 * 5, K])
X_test = np.zeros([6 * 13 * 5, num_features])
X_test_lr = np.zeros([6 * 13 * 5, K])

train_labels = np.hstack((np.ones(5*13), 2*np.ones(5*13), 3*np.ones(5*13), 4*np.ones(5*13), 5*np.ones(5*13), 6*np.ones(5*13)))
test_labels = np.hstack((np.ones(5*13), 2*np.ones(5*13), 3*np.ones(5*13), 4*np.ones(5*13), 5*np.ones(5*13), 6*np.ones(5*13)))

idx_train = 0
idx_test = 0

# for all possible fingerings
for n in range(6):
    
    curr_string_samples = X[n*13*num_rec_per_pos:(n+1)*13*num_rec_per_pos, :]
    curr_string_samp_lr = low_rank_X[n*13*num_rec_per_pos:(n+1)*13*num_rec_per_pos, :]
    
    for m in range(13):
        
        curr_pos_samples = curr_string_samples[m * num_rec_per_pos:(m + 1) * num_rec_per_pos, :]
        curr_pos_samp_lr = curr_string_samp_lr[m * num_rec_per_pos:(m + 1) * num_rec_per_pos, :]
        
        # randomly permute the recording indices
        permuted_indices = np.random.permutation(range(num_rec_per_pos))
        
        # pick the first 5 for training and the remaining 5 for testing
        train_idx = permuted_indices[:5]
        test_idx = permuted_indices[5:]
        
        # store the training feature vectors
        X_train[idx_train:idx_train + 5, :] = curr_pos_samples[train_idx]
        X_train_lr[idx_train:idx_train + 5, :] = curr_pos_samp_lr[train_idx]
        idx_train = idx_train + 5
        
        # store the test feature vectors
        X_test[idx_test:idx_test + 5, :] = curr_pos_samples[test_idx]
        X_test_lr[idx_test:idx_test + 5, :] = curr_pos_samp_lr[test_idx]
        idx_test = idx_test + 5
        
        # store the corresponding indices
        train_indices[n, m] = train_idx
        test_indices[n, m] = test_idx
```

## IV - TESTING THE MODEL

Now, we have at our disposal a trained classifier. Given a test recording, we process it in the same fashion as the training recordings. Once the feature vector has been extracted, we can use our system to predict the class/label of the test recording, i.e the string on which the note has been played. 

Recall that we want to use our knowledge of the guitar geometry to _discard_ any string on which the estimated fundamental cannot be played. Instead of doing a "hard classification", we retrieve the __classes probabilities__, that is the probabilities that the current test example belongs to each class. 
The probability of each "impossible class" is set to 0, so as to force our final decision towards a plausible position for the current fundamental. Among the remaining plausible classes, we pick the one with the largest probability.

### IV.I - TESTING A SINGLE RECORDING

In this subsection, we test our model on a single, hand-picked recording in order to visualize what happens.

```python
# test recording parameters
str_num = 6
fret_num = 5
take_num = int(test_indices[str_num - 1, fret_num][1]) + 1

file_folder = 'string_' + str(str_num) + '/string_' + str(str_num) + '_fret_' + str(fret_num) + '/'
file_name = 'string_' + str(str_num) + '_fret_' + str(fret_num) + '_take_' + str(take_num) + '.wav'

# read the recording
fs_recording, current_recording = wavfile.read(PATH_RECORDINGS + file_folder + file_name)
y = current_recording / max(current_recording)

# extract the first 1.5 seconds of the recording
y = y[:int(1.5*fs_recording)]

# TMP: initializations and parameters
tot_num_samp = y.shape[0]
num_samp_per_frame = int(fs_recording * frame_duration)
window = np.hamming(num_samp_per_frame)
hop_size = int(num_samp_per_frame / 2)
num_frames = int((tot_num_samp - num_samp_per_frame) / hop_size)

# estimate the fundamental frequency
estimated_fundamental = estimate_fundamental(np.abs(np.fft.rfft(y * np.hamming(tot_num_samp), NFFT)), fs_recording, NFFT, tol, max_num_peaks)
print("\nEstimated fundamental [Hz]: {est_fund}".format(est_fund=estimated_fundamental))

# track the partials in successive time frames with our modified Goertzel
estimated_partials, partials_amplitudes = partials_tracking(y, fs_recording, frame_duration, estimated_fundamental, max_mode, \
                                                   delta_f, 'goertzel')

# extract all the audio features from the current test recording
feature_vector = extract_audio_features(y, fs_recording, estimated_partials, partials_amplitudes, max_mode, num_frames, 10) 
print("\nOriginal feature vector: {feat_vec}".format(feat_vec=feature_vector))

# if dimensionality reduction is used
#feature_vector = feature_vector @ V[:, :K]
#print("\nShrunk: {feat_vec}".format(feat_vec=feature_vector))

# extract the classes probabilities for the current feature vector
classes_prob = clf.predict_proba(feature_vector.reshape(1, -1))
print("\nOriginal classes probabilities (SVM): {class_prob}".format(class_prob=classes_prob))

# search the potential fingerings of the estimated fundamental 
# in our dictionary (fundamental to fingerings mapping)
if estimated_fundamental in fundamental_to_fingerings.keys():
        fingerings_candidates = fundamental_to_fingerings[estimated_fundamental]
else:
    print('Error: Fundamental not found!')
print("\nPotential fingerings: {fing_cand}".format(fing_cand=fingerings_candidates))

# "plausibility filter": only consider the classes (strings)
# where the note can actually be played
string_candidates = []
class_prob_candidates = []
for fingering in fingerings_candidates:
    string_num, fret_pos = fingering
    string_candidates.append(string_num)
    class_prob_candidates.append(classes_prob[0, string_num - 1])

best_candidate = string_candidates[np.argmax(class_prob_candidates)]
print("\n\n\tBest string candidate: {best}".format(best=best_candidate))

if(best_candidate != str_num):
    print("\n\tERROR!")
else:
    print("\n\tFingering correctly estimated!")
```

### IV.II TESTING ALL THE RECORDINGS

We can now test our system on all the test recordings.

```python
# initializations
num_errors = 0

print("\nTesting our model...")

for n in range(5, 6):
    
    print("\nString n°{str_num}".format(str_num=n+1))
    
    for m in range(5, 13):
        
        print("\n\tFret n°{fret_num}".format(fret_num=m))
        
        # initializations for the current fingering
        file_folder = 'string_' + str(n + 1) + '/string_' + str(n + 1) + '_fret_' + str(m) + '/'
        file_name_prefix = 'string_' + str(n + 1) + '_fret_' + str(m) + '_take_'
        
        curr_test_idx = test_indices[n, m]
        
        # We evaluate one test recording at a time
        for rec_num in curr_test_idx:
            
            # read the recording from the data folder
            file_name = file_name_prefix + str(int(rec_num + 1)) + '.wav'            
            fs_recording, current_recording = wavfile.read(PATH_RECORDINGS + file_folder + file_name)
            y = current_recording / max(current_recording)
            
            # extract the first 1.5 seconds of the recording
            y = y[:int(1.5*fs_recording)]
            
            # TMP: PARAMETERS (TO PUT SOMEWHERE ELSE)
            tot_num_samp = y.shape[0]
            num_samp_per_frame = int(fs_recording * frame_duration)
            window = np.hamming(num_samp_per_frame)
            hop_size = int(num_samp_per_frame / 2)
            num_frames = int((tot_num_samp - num_samp_per_frame) / hop_size)
            
            # estimate the fundamental frequency
            estimated_fundamental = estimate_fundamental(np.abs(np.fft.rfft(y * np.hamming(tot_num_samp), NFFT)), fs_recording, \
                                                 NFFT, tol, max_num_peaks)
            # TMP: PROBLEM WITH THE FIRST STRING
            if(n == 0 and m == 0):
                estimated_fundamental = 329.6
            print("\n\t\tEstimated fundamental [Hz]: {est_fund}".format(est_fund=estimated_fundamental))
            
            # track the partials in successive time frames with our modified Goertzel
            estimated_partials, partials_amplitudes = partials_tracking(y, fs_recording, frame_duration, estimated_fundamental, max_mode, \
                                                   delta_f, 'goertzel')
            
            # extract all the audio features from the current test recording
            feature_vector = extract_audio_features(y, fs_recording, estimated_partials, partials_amplitudes, \
                                                    max_mode, num_frames, 10)
            print("\n\t\tOriginal feature vector: {feat_vec}".format(feat_vec=feature_vector))
            
            # apply PCA to keep significant features
            #feature_vector = feature_vector @ V[:, :K]
            #print("\n\t\tShrunk feature vector: {feat_vec}".format(feat_vec=feature_vector))
            
            # extract the classes probabilities for the current feature vector
            classes_prob = clf.predict_proba(feature_vector.reshape(1, -1))
            print("\n\t\tOriginal classes probabilities: {class_prob}".format(class_prob=classes_prob))
            
            # search the potential fingerings of the estimated fundamental 
            # in our dictionary (fundamental to fingerings mapping)
            if estimated_fundamental in fundamental_to_fingerings.keys():
                fingerings_candidates = fundamental_to_fingerings[estimated_fundamental]
            else:
                print('Error: Fundamental not found!')
            print("\n\t\tPotential fingerings: {fing_cand}".format(fing_cand=fingerings_candidates))
            
            # "plausibility filter": only consider the classes (strings)
            # where the note can actually be played
            string_candidates = []
            class_prob_candidates = []
            for fingering in fingerings_candidates:
                string_num, fret_pos = fingering
                string_candidates.append(string_num)
                class_prob_candidates.append(classes_prob[0, string_num - 1])
            
            # select the best candidate, i.e the class with the highest probability
            best_candidate = string_candidates[np.argmax(class_prob_candidates)]
            print("\n\t\tBest string candidate: {best}".format(best=best_candidate))
            
            # print the result of the classification
            if(best_candidate != (n + 1)):
                num_errors = num_errors + 1
                print("\n\t\tERROR!")
            else:
                print("\n\t\tFingering correctly estimated!")
```
