# EXPLORING SOME AUDIO FEATURES



In a previous notebook, we have been able to build a simple classifier solely based on the inharmonicity profile of the different strings. As we have seen, this single-feature classifier performance was mitigated. In some cases, the distance between the potential inharmonicities was too small for our method to handle. Thus, we are looking for a way to improve our classification scheme. 


The idea is now to explore other __audio features__ that might be used to distinguish different strings or fingerings. Our choices are based on previous work on the subject and mainly inspired by Jakob Abesser, _Automatic String Detection for Bass Guitar and Electric Guitar_. As we have seen, if the detection of the inharmonicity coefficient is off (in particular, if it is negative), the classification fails. Therefore, it makes sense to use other features in order to build a more robust classifier.


__Which audio features?__ Potential features include the followings:


- The __general time envelope__ of the recording, obtained by fitting an exponentially decaying envelope.


- The __relative amplitudes__ of the partials, compared to the fundamental amplitude.


- The __decay times__ of the partials amplitudes.


- The __inharmonicity coefficient__, as discussed in previous notebooks.


__Why these features?__ The choice of such features is fairly intuitive. Different strings (different materials (bronze vs. steel), different thickness (lower vs. higher strings), etc...) might attenuate different partials differently, which would yield different relative amplitudes and decay times. Similarly, we have seen that the inharmonicity coefficient is a distinctive feature of the different strings of the guitar. The time envelope idea comes from a simple physical observation: fretted notes tend to be "sharper"/"narrower" than open notes, suggesting also a difference in time.


__Approach:__ We will try to automatize the extraction of such features for a given recording. We will use all the tools developed so far, in particular our frequency ratios method to extract the fundamental, and our modified Goertzel's method to track the partials (and their amplitude) in successive time frames. The ultimate goal is to form a __feature vector__ that describes well the string on which the note is being played, in an attempt to improve our classification system.

```python
# imports
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import numpy.matlib

from scipy.io import wavfile

from MUSIC import *
from spectral_grid_search import *
from pitch_estimation import *
from audio_feature_extraction import *
```

## 0 - PARAMETERS

```python
# general parameters
max_mode = 8
modes_num = np.arange(1, max_mode + 1)

# algorithms parameters
NFFT = 2**13
tol = .05
max_num_peaks = 10

delta_f = 40

# recording parameters
PATH_RECORDINGS = '../recordings/string_6/string_6_fret_0/'
file_name = 'string_6_fret_0_take_1.wav'

# read the recording from the folder
fs_recording, current_recording = wavfile.read(PATH_RECORDINGS + file_name)
y = current_recording / max(np.abs(current_recording))

# resize: we consider the first 2.0 seconds of each recording
recording_duration = 1.5
y = y[:int(recording_duration*fs_recording)]

# time frames' parameters
tot_num_samp = y.shape[0]
frame_duration = .75
num_samp_per_frame = int(fs_recording * frame_duration)
window = np.hamming(num_samp_per_frame)
hop_size = int(num_samp_per_frame / 2)
num_frames = int((tot_num_samp - num_samp_per_frame) / hop_size)
```

## I - TRACKING THE PARTIALS AND THEIR AMPLITUDES

Using our toolbox of spectral estimation methods, we track the partials and their amplitudes in successive time frames. These will be used later on in the computation of our audio features.

```python
# parameters & initializations
estimated_partials = np.zeros([max_mode, num_frames])
partials_amplitudes = np.zeros([max_mode, num_frames])
norm_freq_diff_recording = np.zeros([max_mode, num_frames])

# estimate the fundamental frequency
estimated_fundamental = estimate_fundamental(np.abs(np.fft.rfft(np.hamming(y.shape[0]) * y, NFFT)), fs_recording, \
                                                 NFFT, tol, max_num_peaks)
print("\nEstimated fundamental [Hz]: {est_fund}".format(est_fund=estimated_fundamental))

# analysis of successive frames
freq_ranges = [(int(k*estimated_fundamental - delta_f), int(k*estimated_fundamental + delta_f)) \
               for k in range(1, max_mode + 1)]

for n in range(num_frames):
    
    # extract the current time frame (windowed block)
    curr_frame = y[n*hop_size:n*hop_size + num_samp_per_frame] * window
    
    # estimate the partials using our modified Goertzel's algorithm
    estimated_partials[:, n], partials_amplitudes[:, n] = goertzel(curr_frame, fs_recording, frame_duration, freq_ranges)
    
    # print status of the current frame
    print("\nFrame nÂ°{frame_num}".format(frame_num=n+1))
    print("\nEstimated partials [Hz]: {est_part}".format(est_part=estimated_partials[:, n]))
    
    # OPTIONAL: visualize the frequency deviation/inharmonicity curves
    norm_freq_diff_recording[:, n] = (estimated_partials[:, n] - (estimated_partials[0, n] * modes_num))/(estimated_partials[0, n])
    plt.plot(modes_num, norm_freq_diff_recording[:, n], 'gx-')
    plt.plot(modes_num, np.zeros(max_mode), 'k--')
    plt.xlabel('Partial number')
    plt.ylabel('Normalized frequency difference')
    plt.title('INHARMONICITY CURVE')
    plt.grid()
    plt.tight_layout()
    plt.show()
```

## II - EXTRACTING THE AUDIO FEATURES FROM THE RECORDINGS

Once the partials' frequencies and amplitudes have been estimated, we can extract the audio features described above. The overall time envelope is captured by fitting an exponential decaying envelope. The relative amplitudes are obtained as simple ratios. We then compute the decay times by fitting a degree-1 polynomial on the logarithm of the amplitudes. Finally, the inharmonicity are obtained using our 2D least squares.

```python
# 1] extract the general time envelope by fitting an exponential decay curve

# divide the signal into blocks and take the maxium (in absolute value) over each block
n = 10
start_point_envelope = np.argmax(np.abs(y))
t = np.linspace(start_point_envelope/fs_recording, recording_duration, int(fs_recording*recording_duration - start_point_envelope))
nblocks = int(np.floor(t.shape[0]/n))
trun = np.mean(t[:n*nblocks].reshape([n, nblocks]), 1)
envelope = np.max(y[start_point_envelope:n*nblocks + start_point_envelope].reshape([n, nblocks]), 1)

# fit an exponential decaying curve (polynomial fit on the log of the envelope)
poly_coef = np.polyfit(trun, np.log(envelope), 1)
global_decay_time = -1/poly_coef[0]
global_lead_const = np.exp(poly_coef[1])

# visualization
plt.figure()
plt.plot(t, np.abs(y[start_point_envelope:]))
plt.plot(t, global_lead_const * np.exp(- t / global_decay_time), 'r--')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend(['Rec.', 'Fit'])
plt.title('Global time envelope')
plt.grid()
plt.tight_layout()
plt.show()

print("\nGlobal decay time: {glob_decay_time}".format(glob_decay_time=global_decay_time))

# 2] extract the partials' relative amplitudes w.r.t to the fundamental
rel_part_amp = partials_amplitudes / partials_amplitudes[0, :]

# visualization
plt.figure()
plt.plot(modes_num, rel_part_amp)
plt.xlabel('Partial number')
plt.ylabel('Relative amplitude')
plt.title('Partials relative amplitudes')
plt.grid()
plt.tight_layout()
plt.show()

# 3] compute approximate decay times (and lead. constants) of the partials' amplitudes
decay_times = np.zeros(max_mode)
lead_constants = np.zeros(max_mode)

# for each partial in some predefined range, we capture the temporal evolution 
# of the partial's amplitudes by fitting an exponential decaying envelope
for partial_num in range(max_mode):
    
    # extract the current partial amplitudes in the successive frames
    curr_part_amp = partials_amplitudes[partial_num, :]
    
    # normalize to avoid numerical errors/overflows
    norm_curr_part_amp = curr_part_amp / max(curr_part_amp)
    
    # quick exponential fit: polynomial fit on the log
    curr_poly_coef = np.polyfit(np.arange(1, num_frames + 1), np.log(norm_curr_part_amp), 1)
    curr_decay_time = -1/curr_poly_coef[0]
    curr_lead_const = np.exp(curr_poly_coef[1])
    
    decay_times[partial_num] = curr_decay_time
    lead_constants[partial_num] = curr_lead_const
    
    print("\nPartial nÂ°{part_num}".format(part_num=partial_num+1))
    
    # visualization
    plt.figure()
    plt.plot(np.arange(1, num_frames + 1), norm_curr_part_amp, 'bx-')
    plt.plot(np.arange(1, num_frames + 1), curr_lead_const * np.exp(- np.arange(1, num_frames + 1) / curr_decay_time), 'ro-')
    plt.xlabel('Time / Frame number')
    plt.ylabel('Normalized partial amplitude')
    plt.title('Exponential fit')
    plt.legend(['DATA', 'FIT'])
    plt.grid()
    plt.tight_layout()
    plt.show()
    

# 4] estimate the inharmonicity coefficient
# method 1 (default): 2D Least squares (preferred)
partials_concat = [(estimated_partials[:, n]/estimated_partials[0, n])**2 for n in range(num_frames)]
LHS = np.hstack(partials_concat)
system_matrix = np.matlib.repmat(np.vstack((np.arange(1, max_mode + 1)**2, \
                                            np.arange(1, max_mode + 1)**4)).T, num_frames, 1)
coef, _, _, _ = np.linalg.lstsq(system_matrix, LHS)
b_hat = coef[1]
print("\nInharmonicity coefficient: {b_hat}".format(b_hat=b_hat))

# alternative: 1D Least squares (average)
partials_concat = [(((estimated_partials[1:, n]/estimated_partials[0, n])**2 - modes_num[1:]**2)/modes_num[1:]**4) for n in range(num_frames)]
print("\t(alternative method: {b_hat_2})".format(b_hat_2=np.mean(partials_concat)))
```

```python
print("\nSummary - Extracted features")

print("\nGlobal decay time: {glob_decay_time}".format(glob_decay_time=global_decay_time))

print("\nRelative partial amplitudes: {rel_part_amp}".format(rel_part_amp=rel_part_amp))

print("\nPartials' amplitudes decay times: {part_decay_times}".format(part_decay_times=decay_times))

print("\nInharmonicity coefficient: {b_hat}\n".format(b_hat=b_hat))
```

## III - FORMING THE FEATURE VECTOR


The resulting feature vector is simply obtained by stacking all the extracted audio features into a single vector for further use in a classification pipeline.

```python
# stack all the features in a feature vector
feature_vector = np.hstack((global_decay_time, rel_part_amp.T.reshape([max_mode * num_frames]), decay_times, b_hat))
print("Feature vector: {feat_vec}".format(feat_vec=feature_vector))
```
