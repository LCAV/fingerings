# MODELS COMPARISON
In this notebook, we compare two simple models for the guitar: the string motion & the transverse force exerted on the guitar bridge.

```python
%matplotlib inline
import numpy as np
import sounddevice as sd
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy import signal
```

```python
# Standard string parameters
fundamental_freqs = np.array([82, 110, 147, 196, 247, 330])
string_diameters = (1/39.17) * np.array([47, 39, 30, 23, 14, 10]) * 1e-3
string_length = 0.64
p_steel = 7850
p_bronze = 8600
p_strings = p_bronze * (string_diameters/2)**2 * np.pi
p_strings[4:] = p_steel * (string_diameters[4:]/2)**2 * np.pi
tensions = 4*(string_length**2)*(fundamental_freqs**2)*p_strings
velocities = np.sqrt(tensions/p_strings)

# "Playing" parameters
original_pluck = 1/5 + (np.random.rand() - 0.5)/30
init_displacement = 0.002
max_mode = 20
modes_num = np.arange(1, max_mode + 1)

# Sampling & spectrogram parameters
f_samp = 44100
T_samp = 1/f_samp
simulation_duration = 5.0
t = np.arange(0.0, simulation_duration, T_samp)
NFFT = 4096
noverlap = NFFT/2
```

## Model n°1: String motion
Our firt simple model only considers the string itself and its motion as a function of time. Given an initial deflection, a plucking point, and a string, we can compute the composite string motion by summing up all the modes with the appropriate amplitudes. The drawback is that we need to select a point along the string to compute the string motion at that particular point, which changes a little bit the resulting vibration.

```python
def create_string_motion(modes_num, max_mode, init_displacement, plucking_point, ref_point, f, c, t):
    
    """ Compute the composite string motion for a given excitation.
    
        input:
            - modes_num: the modes serial numbers
            - max_mode: the maximum mode number to consider
            - init_displacement: the initial deflection from equilibrium
            - plucking_point: the plucking point on the string (fraction of its length)
            - ref_point: the reference point where the motion is to be computed
            - f: the fundamental frequency of the string
            - c: the wave velocity in the string
            - t: the time indices
            
        output:
            - string_motion: the composite string motion
    """
    
    # Compute the mode amplitudes, depending on the initial displacement and plucking point
    mode_amplitudes = ((1/plucking_point)**2 * init_displacement)/(2 * modes_num**2 * np.pi**2) * \
        np.sin(modes_num * np.pi * plucking_point)
    
    # Compute the composite string motion at the referece point by superimposing 
    # the contributions from all the modes, and accounting for the damping effects
    # (exponential decay in amplitude)
    string_motion = np.zeros(t.shape)
    for n in np.arange(max_mode):
        string_motion += mode_amplitudes[n] * np.sin(2 * np.pi * (f*(n + 1) * t)) * \
            np.sin((2 * np.pi * f *(n + 1) * ref_point)/c) * np.exp(- 2 * np.pi**2 * np.sqrt(f*(n + 1)) * 1e-5 * t)
    
    return string_motion
```

## Model n°2: Transverse force exerted on the bridge
With this model, we consider the transverse force exerted on the guitar bridge. This force depends on the string motion, in that it oscillates with somewhat the same frequencies. However, since the bridge is at a fixed location, there is no need to select a point on the string. Thus, for a given deflection, plucking point and string, the output is the same.

```python
def compute_bridge_force(modes_num, max_mode, init_displacement, plucking_point, string_length, \
                         tension, f, c, t):
    
    """ Compute the transverse force exerted by the string on the guitar bridge. 
    
        input:
            - modes_num: the modes serial numbers
            - max_mode: the maximum mode number to consider
            - init_displacement: the initial deflection from equilibrium
            - plucking_point: the point where the string is plucked
            - string_length: the total length of the string
            - tensions: the tension applied on the string
            - f: the fundamental frequency of the string
            - c: wave velocity in the string
            - t: time indices
            
        output:
            - trans_bridge_force: the total force exerted on the bridge
    """
    
    # compute the amplitude of each considered harmonic
    harmonics_amplitude = (2*init_displacement*tension/(modes_num * np.pi * string_length)) \
    * (np.sin(modes_num*plucking_point*np.pi)/(plucking_point * (1 - plucking_point)))
    
    # compute the transverse force exerted on the bridge by summing up the contributions
    # from all vibrational modes
    trans_bridge_force = np.zeros(t.shape)
    for n in np.arange(max_mode):
        trans_bridge_force += harmonics_amplitude[n] * np.cos(2 * np.pi * f * (n + 1) * t) * \
            np.exp(- 2 * np.pi**2 * (f*(n + 1)) * 1e-5 * t)
    
    # normalization
    trans_bridge_force /= max(trans_bridge_force)
    
    return trans_bridge_force
```

## Comparison
We now compare the two models

```python
# Considered string & parameters
string_num = 1
tension = tensions[string_num - 1]
velocity = velocities[string_num - 1]
f = fundamental_freqs[string_num - 1]
ref_point = string_length * 1/4
```

```python
# MODEL 1
string_motion = create_string_motion(modes_num, max_mode, init_displacement, original_pluck, ref_point, \
                                     f, velocity, t)
string_motion /= max(string_motion)
modes_amplitude = ((1/original_pluck)**2*init_displacement)/(2*modes_num**2*np.pi**2) * \
        np.sin(modes_num*np.pi*original_pluck)
    
# MODEL 2
trans_bridge_force = compute_bridge_force(modes_num, max_mode, init_displacement, original_pluck, \
                                          string_length, tension, f, velocity, t)
harmonics_amplitude = (2*init_displacement*tension/(modes_num * np.pi * string_length)) \
    * (np.sin(modes_num*original_pluck*np.pi)/(original_pluck * (1 - original_pluck)))
```

```python
# plots & visualization
fig = plt.figure(figsize=(20, 12))

ax1 = plt.subplot2grid((3, 2), (0, 0), rowspan=1, colspan=1)
ax1.stem(harmonics_amplitude)
plt.xlabel('harmonics serial number n')
plt.ylabel('amplitude Fn')
plt.title('Transverse bridge force: modes amplitude')

ax2 = plt.subplot2grid((3, 2), (1, 0), rowspan=1, colspan=1)
ax2.plot(trans_bridge_force[0:3000])
plt.xlabel('samples')
plt.ylabel('amplitude')
plt.title('Transverse bridge force: waveform')

ax3 = plt.subplot2grid((3, 2), (2, 0), rowspan=1, colspan=1)
ax3.plot(np.arange(200)*f_samp/NFFT, 20 * np.log10(np.abs(np.fft.fft(trans_bridge_force, NFFT))[0:200]))
plt.xlabel('frequency [Hz]')
plt.ylabel('magnitude [dB]')
plt.title('Transverse bridge force: spectrum')


ax5 = plt.subplot2grid((3, 2), (0, 1), rowspan=1, colspan=1)
ax5.stem(modes_amplitude, 'r')
plt.xlabel('modes serial number n')
plt.ylabel('amplitude An')
plt.title('String motion: modes amplitude')

ax6 = plt.subplot2grid((3, 2), (1, 1), rowspan=1, colspan=1)
ax6.plot(string_motion[0:3000], 'r')
plt.xlabel('samples')
plt.ylabel('amplitude')
plt.title('String motion: waveform')

ax7 = plt.subplot2grid((3, 2), (2, 1), rowspan=1, colspan=1)
ax7.plot(np.arange(200)*f_samp/NFFT, 20 * np.log10(np.abs(np.fft.fft(string_motion, NFFT))[0:200]), 'r')
plt.xlabel('frequency [Hz]')
plt.ylabel('magnitude [dB]')
plt.title('String motion: spectrum')

plt.tight_layout()
plt.show()
```
