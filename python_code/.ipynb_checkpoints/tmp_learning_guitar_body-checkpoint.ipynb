# LEARNING THE EFFECT OF THE GUITAR BODY

In this notebook, we try to estimate the effect of the guitar body using linear regression and learning techniques. The input to our system is our simple model output, i.e the string motion (see string_simulation notebook) and the target/output is a real recording. We will assume that the guitar body acts as an FIR filter with K coefficients and will try to match the string motion to the real recording by finding a suitable filter.

```python
%matplotlib inline
import numpy as np
import sounddevice as sd
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy import signal
```

##### GLOBAL PARAMETERS

```python
# Standard string parameters
fundamental_freqs = np.array([82, 110, 147, 196, 247, 330])
string_diameters = (1/39.17) * np.array([47, 39, 30, 23, 14, 10]) * 1e-3
string_length = 0.64
p_steel = 7850
p_bronze = 8600
p_strings = p_bronze * (string_diameters/2)**2 * np.pi
p_strings[4:] = p_steel * (string_diameters[4:]/2)**2 * np.pi
tensions = 4*(string_length**2)*(fundamental_freqs**2)*p_strings
velocities = np.sqrt(tensions/p_strings)

# "Playing" parameters
original_pluck = 1/5 + (np.random.rand() - 0.5)/20
init_displacement = 0.002
max_mode = 20
modes_num = np.arange(1, max_mode + 1)

# Sampling & spectrogram parameters
f_samp = 44100
T_samp = 1/f_samp
simulation_duration = 5.0
t = np.arange(0.0, simulation_duration, T_samp)
NFFT = 4096
noverlap = NFFT/2
```

```python
def create_string_motion(modes_num, max_mode, init_displacement, plucking_point, f, c, t):
    
    # Compute the mode amplitudes, depending on the initial displacement and plucking point
    # (NOTE: all are scaled to be able to hear the sound properly)
    mode_amplitudes = 200 * ((1/plucking_point)**2*init_displacement)/(2*modes_num**2*np.pi**2) * \
        np.sin(modes_num*np.pi*plucking_point)
    
    # "Reference" point along the string
    #x = 0       # nut
    x = 0.64    # bridge/saddle
    #x = 0.64/2  # middle of the neck
    #x = 0.64 * (1/5 + 1/4)/2 # around the soundhole
    
    # Compute the composite string motion at the referece point by superimposing 
    # the contributions from all the modes, and accounting for the damping effects
    # (exponential decay in amplitude)
    string_motion = np.zeros(t.shape)
    for n in np.arange(max_mode):
        string_motion += mode_amplitudes[n] * np.sin(2 * np.pi * (f*(n + 1) * t)) * \
            np.sin((2 * np.pi * f *(n + 1) * x)/c) * np.exp(- 1/np.sqrt(f * (n + 1)) * 10 * t)
    
    return string_motion
```

```python
def estimate_start_point(target_signal, ref_signal):
    
    # window size for the correlation-based detection
    corr_window_size = np.min(np.where(target_signal > 0.5)) + 1000

    # correlation-based detection of the starting point of our recording
    starting_point = np.argmax(np.correlate(target_signal[0:corr_window_size], \
                                            ref_signal[0:corr_window_size], 'full')) - corr_window_size
    
    return starting_point
```

```python
# input signal (model output)
string_num = 1
f = fundamental_freqs[string_num - 1]
c = velocities[string_num - 1]
string_motion = create_string_motion(modes_num, max_mode, init_displacement, original_pluck, f, c, t)
string_motion /= max(string_motion)

# parameters
N = 5000
K = 1000
num_recordings = 1

PATH_RECORDINGS = '../recordings/open_strings/open_low_E/'

# initialize the system matrix ("convolution matrix")
X = np.zeros([num_recordings * N, K + 1])
y = np.zeros(num_recordings * N)

# fill in the system matrix ("convolution matrix")
for i in range(num_recordings):
    
    # read the recording
    file_name = 'open_low_E_take_' + str(i + 1) + '.wav'
    fs_recording, current_recording = wavfile.read(PATH_RECORDINGS + file_name)
    
    # estimate the starting point
    starting_point = estimate_start_point(current_recording/max(current_recording), string_motion)
    
    plt.figure()
    plt.plot(current_recording[starting_point:starting_point + 500])
    plt.show()
    
    plt.figure()
    plt.plot(string_motion[0:500])
    plt.show()
    
    # target signal (normalized) from starting point
    y[i * N:(i + 1) * N] = (current_recording/max(current_recording))[starting_point:N + starting_point]
    
    for j in range(N):
        
        if(j <= K):
            X[j + i * N] = np.pad(string_motion[j::-1], (0, K - j), 'constant')
        else:
            X[j + i * N] = string_motion[j:j - (K + 1):-1]

# Least Squares solution
w_hat = np.linalg.solve(X.T @ X, X.T @ y)
```

```python
plt.figure()
plt.stem(w_hat)
print(w_hat)
```

```python
test = signal.convolve(string_motion, w_hat)
plt.figure()
plt.plot(test[500:3000])
```
