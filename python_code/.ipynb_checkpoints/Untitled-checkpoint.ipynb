```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
n = np.arange(100)
B = 1e-4
f0 = 82.4

fn = n*f0*np.sqrt(1+B*n**2)

plt.figure()

plt.plot(n, n*f0, 'kx-', label='Ideal')
plt.plot(n, fn, 'rx-', label='Stiff')

plt.xlabel('Partial number')
plt.ylabel('Frequency [Hz]')
plt.title('Effect of string stiffness')

plt.legend(loc = 'best')

plt.tight_layout()

plt.savefig('stiffness.png')

plt.show()
```

```python
t = np.linspace(0.0, 5.0, 5.0 * 8e3)
x = np.sin(2*np.pi*10*t) * np.exp(-t)

plt.figure()
plt.plot(t, x)
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.title('Damped sinusoidal oscillation')

plt.tight_layout()

plt.savefig('damping.png')

plt.show()
```

```python
modes_num = np.arange(1, 11)
plucking_point_1 = 1/2
plucking_point_2 = 1/5
init_displacement = 3e-3

mode_amplitudes_1 = ((1/plucking_point_1)**2*init_displacement)/(2*modes_num**2*np.pi**2) * \
        np.sin(modes_num*np.pi*plucking_point_1)
    
mode_amplitudes_2 = ((1/plucking_point_2)**2*init_displacement)/(2*modes_num**2*np.pi**2) * \
        np.sin(modes_num*np.pi*plucking_point_2)

plt.figure(figsize=(10, 8))

ax1 = plt.subplot2grid((2, 2), (0, 0), rowspan=1, colspan=1)
ax1.stem(modes_num, np.abs(mode_amplitudes_1/max(mode_amplitudes_1)), 'k')
ax1.plot(np.arange(0, 12), np.zeros(12), 'k-')
plt.ylim(-.2, 1.2)
plt.xlim(0, 11)
plt.xlabel('Mode number')
plt.ylabel('Relative amplitude')
plt.title('Relative mode amplitude, plucked at 1/2')

ax2 = plt.subplot2grid((2, 2), (0, 1), rowspan=1, colspan=1)
ax2.stem(modes_num, np.abs(mode_amplitudes_2/max(mode_amplitudes_2)), 'k')
ax2.plot(np.arange(0, 12), np.zeros(12), 'k-')
plt.ylim(-.2, 1.2)
plt.xlim(0, 11)
plt.xlabel('Mode number')
plt.ylabel('Relative amplitude')
plt.title('Relative mode amplitude, plucked at 1/5')

plt.tight_layout()
plt.show()
```
