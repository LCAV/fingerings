## II. THEORY AND BACKGROUND: THE PHYSICS BEHIND THE GUITAR <br />

<div style="text-align:justify;text-indent:15px;"> In this introductory chapter, we will present several theoretical results on vibrating strings and the guitar, in an effort to understand the physics behind the sounds of the guitar. When a guitarist plucks a string on a guitar, a chain of several oscillators or vibrators are excited. The complex tones that reach our ears are thus the result of multiple systems that interact together. In the following, we present these oscillating systems and discuss some of their properties that might be of interest to us in the fingering estimation task. </div>

### II.I Vibrating strings <br />

<div style="text-align:justify;"> The first oscillator involved in the production of a guitar sound is the guitar string. It will play a prominent role throughout our work. A vibrating string can be viewed as a limiting case of a mass-spring system (a chain-like collection of springs and masses). A mass-spring system exhibits normal or characteristic modes of vibration. These are motion patterns in which the system moves sinusoidally at a natural or resonance frequency. As the number of masses and springs increases, the system shows a "wavelike" behavior and resembles a vibrating string with a uniformly distributed mass ([2]). </div>

<br />

#### II.I.I Modes of vibration of an ideal string <br />

<div style="text-align:justify;"> On an ideal vibrating string, transverse waves satisfy the wave equation ([2]) </div><br />
\begin{equation}
    \frac{\partial^2 y}{\partial t^2} = \frac{T}{\mu} \frac{\partial^2 y}{\partial x^2},
\end{equation}
<br /><div style="text-align:justify;"> where $y$ is the string motion in the transverse plane, $x$ is the position along the string, $T$ is the tension applied to the string, and $\mu$ is the linear density of the string. The wave will travel at a speed $c = \sqrt{T/\mu}$. </div>

<div style="text-align:justify;"> On a guitar, the two ends of any string are fixed. For instance, on an acoustic guitar, one end is attached to the tuning mechanisms while the other is usually pinned to the guitar bridge. These boundary conditions affect the way waves are reflected. If we consider a string of length $L$ fixed at both ends (at $x = 0$ and $x = L$), the boundary conditions restrict the possible frequencies to the discrete frequencies </div><br />

\begin{equation}
    f_n = n \ \frac{c}{2L} = n \ f_1.
\end{equation}

<br /><div style="text-align:justify;"> These frequencies correspond to integer multiples of a fundamental frequency $f_1 = c/2L$. Therefore, the string exhibits harmonic modes of vibration. The general solution of the ideal wave equation above can then be expressed as a sum of these normal modes, </div><br />

\begin{equation}
    y(x, t) = \sum_{n} y_n(x, t) = \sum_{n} C_n \sin(\omega_n t + \phi_n) \sin(k_n x)
\end{equation}

<br /><div style="text-align:justify;"> The waves propagating on the string will create both constructive and destructive interference, which will create standing or stationary waves. The situation is well captured by the following small animation, where the first four vibrational modes of an ideal string are depicted. </div>

<img src="images_animations/vibrational_modes.gif", width=250, height=200>

#### II.I.II "Vibration recipes" <br />

<div style="text-align:justify;"> When a guitar string is plucked, pulse waves propagate in both directions from the plucking point. When these pulses reach the fixed ends of the string, they reflect back and create standing waves. The resulting string motion is a combination of the normal vibrational modes of the string, as we have seen in the previous subsection. The "vibration recipe", a term that denotes which modes are present and their amplitudes ([1]), depends on the plucking position. Depending on where the string is plucked, some modes will be (nearly) absent from the recipe. Benade states in [1] that "the excitation of a mode is proportional to the size of the mode's excursion at the plucking point". For instance, if the string is plucked exactly at its center ($x = L/2$), the even-numbered harmonics will vanish from the recipe since any even-numbered mode has a node at this particular point. Typically, a guitarist plucks the strings at approximately 1/4 or 1/5 of their lengths, though one can vary the plucking point to change the resulting sound. Benade also affirms that the amplitude of the $n$-th mode is proportional to its "serial number" via $1/n^2$, implying that higher modes are progressively less excited ([1]). </div><br />

<img src="images_animations/plucking.png">

<br /><div style="text-align:justify;"> Additionally, the way of plucking (either with a plectrum or with the fingers, and with more or less strength) also affects the vibration recipe. The nature of the excitation influences the initial amplitudes of the various modes making up the overall string motion, and changing the resulting sound. </div>

#### II.I.III String stiffness and damped oscillations <br />

<div style="text-align:justify;"> Until here, our analysis has been done for ideal, flexible strings. However, real strings are not completely flexible, but rather stiff. To account for string stiffness, Fletcher and Rossing ([2]) define an inharmonicity constant $B$ as </div><br />

\begin{equation}
    B = \frac{\pi^2 E S K^2}{T L^2},
\end{equation}

<br /><div style="text-align:justify;">where $E$ is Young's modulus, $S$ is the cross-sectional area of the string, $K$ is the radius of gyration, $T$ is the applied tension, and $L$ the string length. This constant depends both on the "setting" (applied tension and string length) and more crucially on the material (through the elastic modulus and the string-related quantities). The effect of string stiffness on the frequency of the harmonics can be summarized by the relation ([2]) </div><br />

\begin{equation}
    f_n \approx n \ f^{ideal}_1 \ \sqrt{1 + B n^2}.
\end{equation}

<br /><div style="text-align:justify;"> From the above formula, we see that $f_n$, the characteristic frequency of the $n$-th mode, is no longer strictly harmonic. Indeed, it is no longer an integer multiple of the fundamental frequency, but rather slightly increased, hence the name "inharmonicity". We will talk about partials instead of harmonics to distinguish between the two, even though both terms are sometimes used interchangeably. The following plot illustrates the effect of string stiffness and inharmonicity on the frequency of the partials, and we compare it to ideally flexible strings. </div>

<img src="images_animations/stiffness.png">

<br /><div style="text-align:justify;"> Another important effect to take into account is damping. A real vibrating string cannot oscillate indefinitely. It is subject to three types of damping effects: air damping, internal damping, and energy transfer to other vibrating systems. Air damping lowers the vibrational frequency of the modes very slightly, and it produces an exponential decay in amplitude. The figure below shows an example of a general damped oscillation with an exponential decay in amplitude. Internal damping is material-specific (related to elasticity and Young's modulus), and can be neglected for steel strings. The energy transfer of vibrations on a guitar can happen from the strings to the bridge. For steel-string guitars, air damping seems to be dominant ([2]). </div>

### II.II Guitar body <br />

<div style="text-align:justify;"> Another oscillating system on a guitar is the guitar body, which is composed of several parts: the top plate, the back plate, and the air cavity. The top and back plates are two-dimensional oscillators made of wood. They vibrate in certain modes, and these modes strongly depend on several design parameters. At low frequencies, the sound is radiated by the top and back plates, and by the soundhole. The energy is transferred from the top plate to the back plate via both the "ribs" and the air cavity. At higher frequencies, most of the sound is radiated by the top plate ([2]). </div>

#### II.II.I Top and back plates <br />

<div style="text-align:justify;"> The top plate (also referred to as soundboard) has its boundaries more or less fixed (glued to the "ribs"). As we have seen in the case of vibrating strings, the particular boundary conditions affect the characteristic vibrations of the plate: different modes and/or frequencies are observed if the plate has its edges free, clamped or hinged. Typically, the vibrational modes of a guitar top plate are not very different from a clamped/hinged rectangular wooden plate ([2]). We also note that the material has some effect on the vibrational modes' shapes ("wood grain"). </div><br />

<div style="text-align:justify;"> The thickness of the plate also plays a role in the sound produced by a guitar: it can alter the natural frequencies of the sound. Benade ([1]) investigates the effect of loading a small mass on a plate (e.g the bridge). Typically, this tends to lower the frequency of the characteristic modes of the sound. However, it also depends on where the mass is placed: if it is placed near a point of maximum excursion, the effect is maximum, whereas if it is placed near a nodal point or line, it has almost no effect. Similarly, when one cuts out a piece of wood in a plate, as in the case of the soundhole for example, both the mass and the stiffness of the plate are changed. It also depends on where this piece is cut out (point of maximum excursion or nodal point). A luthier  plays with these parameters when he designs a guitar to obtain the frequencies he finds suitable. </div><br />

<div style="text-align:justify;">There is no simple predefined characteristic modes and frequencies for all guitar plates ([1], [2]). It strongly depends on the design and it should be measured individually. </div>

#### II.II.II Air cavity <br />

<div style="text-align:justify;"> The air cavity also acts as an oscillator, with different vibrational modes. These depend partly on the cavity volume and the soundhole diameter, and loosely on the soundhole shape and location ([2]). As in the case of the plates, the vibrational modes depend on the design of the air cavity and they should also be measured individually.
