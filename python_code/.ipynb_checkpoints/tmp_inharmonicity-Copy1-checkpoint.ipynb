# STRING STIFFNESS & INHARMONICITY



In this notebook, we take a closer look at the effect of __string stiffness__ and __inharmonicity__ in distinguishing different strings or fingerings. Theoretical results are taken from Fletcher & Rossing, _The Physics of Musical Instruments_. The "inharmonicity curves" are inspired by T. Kobayashi et al., _Inharmonicity of Guitar String Vibration Influenced by Body Resonance and Fingering Position_. Some elements of inharmonicity estimation are taken from Jakob Abesser, _Automatic String Detection for Bass Guitar and Electric Guitar_.

The effect of string stiffness on the partials' frequencies (assuming _pinned_ ends) can be summarized by the expression

\begin{equation}
    f_n = n \ f_0 \ \sqrt{1 + Bn^2},
\end{equation}

where $f_0$ is the fundamental frequency of the corresponding ideal flexible string, $n$ is the partial (serial) number and $B$ is an __inharmonicity constant or coefficient__ given by
\begin{equation}
    B = \frac{\pi^2 E S K^2}{T L^2}
\end{equation}

In the above equation, $E$ is the elastic modulus, $S$ is the cross-sectional area of the string, $K$ is the radius of gyration (distribution of mass around the axis), $T$ is the applied tension and $L$ the string length.

Theoretically, from the expression above, changing the string (its properties, e.g radius, material, ...) and/or the position on the fretboard (resulting in a different length) will change the inharmonicity coefficient $B$ and the resulting relation between the partials. The name "inharmonicity" comes from the fact that the partials do not appear in a perfect harmonic relation with the fundamental anymore, but are rather slightly "inharmonic" (shifted towards higher frequencies).

Therefore, this phenomenon is of interest to us and might be used to distinguish different strings or fingerings!

```python
# Standard string parameters
fundamental_freqs = np.array([82.4, 110.0, 146.8, 196.0, 246.9, 329.6])
string_diameters = (1/39.17) * np.array([47, 39, 30, 23, 14, 10]) * 1e-3
string_length = 0.64
p_bronze = 8600
p_steel = 7850
p_strings = p_bronze * (string_diameters/2)**2 * np.pi
p_strings[4:] = p_steel * (string_diameters[4:]/2)**2 * np.pi
tensions = 4 * (string_length**2) * (fundamental_freqs**2) * p_strings
velocities = np.sqrt(tensions/p_strings)

max_mode = 8
modes_num = np.arange(1, max_mode + 1)
```

## I - THEORETICAL INHARMONICITY CURVES


We first take a look at theoretical results to better understand what inharmonicity and string stiffness change. To do this, we explore the __frequency difference__ between the partials of an idealized model (where all the partials are perfect harmonics) and those of a more complete model (bringing the string stiffness into the picture). We call these plots "inharmonicity curves" in the following.

```python
# parameters
string_num = 1
fret_pos = 0

current_length = compute_string_length(string_length, fret_pos)
current_velocity = velocities[string_num - 1]
curr_diameter = string_diameters[string_num - 1]
curr_rad_gyr = radii_gyration[string_num - 1]
curr_elastic_mod = elastic_mod[string_num - 1]
curr_tension = tensions[string_num - 1]

# fundamental frequency
f_0 = current_velocity/(2*current_length)

# ===========================
# IDEAL CASE: FLEXIBLE STRING
# ============================
# ideal harmonic partials (perfet integer multiples of the fundamental frequency)
ideal_harmonics = modes_num * f_0
print("Ideal harmonics (w/o stiffness) [Hz]: \n{ideal}".format(ideal=ideal_harmonics))

# =======================
# REAL CASE: STIFF STRING
# ========================
# compute the theoretical inharmonicity coefficient for the given fingering
curr_inharmonicity = (np.pi**3 * curr_elastic_mod * (curr_diameter/2)**2 * curr_rad_gyr**2) \
    / (curr_tension * current_length**2)

# boundary conditions: pinned ends
partials_pinned = modes_num * f_0 * np.sqrt(1 + curr_inharmonicity * modes_num**2) 
print("\nInharmonic partials (w/ stiffness) for PINNED ends [Hz]: \n{part}".format(part=partials_pinned))

# clamped ends (less realistic though)
#partials_clamped = modes_num * f_0 * np.sqrt(1 + curr_inharmonicity * modes_num**2) * \
#    (1 + (2/np.pi)*np.sqrt(curr_inharmonicity) + (4/np.pi**2)*curr_inharmonicity)

#print("\nInharmonic partials (w/ stiffness) for CLAMPED ends [Hz]: \n{part}".format(part=partials_clamped))

# ======================================
# COMPARISON: FLEXIBLE VS. STIFF STRINGS
# =======================================

# measure the distance between ideal harmonics and inharmonic partials: "normalized" frequency difference
norm_freq_diff_pinned = (partials_pinned - ideal_harmonics)/ideal_harmonics
#norm_freq_diff_clamped = (partials_clamped - ideal_harmonics)/ideal_harmonics
#print("\nNormalized frequency difference for PINNED ends: \n{f_diff}".format(f_diff=norm_freq_diff_pinned))
#print("\nNormalized frequency difference for CLAMPED ends: \n{f_diff}".format(f_diff=norm_freq_diff_clamped))

# visualization: partials and inharmonicity curves
plt.figure(figsize = (14, 8))

ax1 = plt.subplot2grid((2, 2), (0, 0), rowspan=1, colspan=1)
ax1.stem(ideal_harmonics, 1/modes_num**2, 'k')
ax1.stem(partials_pinned, 1/modes_num**2, 'g')
#ax1.stem(partials_clamped, 1/modes_num**2, 'b')
plt.xlim(0, 1500)
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude')
plt.title('HARMONICS & PARTIALS')

ax2 = plt.subplot2grid((2, 2), (0, 1), rowspan=1, colspan=1)
ax2.plot(modes_num, norm_freq_diff_pinned, 'go--')
#ax2.plot(modes_num, norm_freq_diff_clamped, 'bd--')
plt.legend(["PINNED", "CLAMPED"])
plt.xlabel('Partial number')
plt.ylabel('Normalized frequency difference')
plt.title('INHARMONICITY CURVE')
plt.grid()
plt.tight_layout()
plt.show()
```

##### OBSERVATIONS: WHAT THE THEORY SAYS


- __Thicker strings are more inharmonic__: By comparing the theoretical inharmonicity curves for all the open strings, we can see that thicker strings are more prone to inharmonicity. The larger cross-sectional areas and radii of gyration make the inharmonicity constant $B$ larger, thus increasing the frequency devations from the ideal harmonics.


- __Fretted notes are more inharmonic than open notes__: For every string, when comparing the inharmonicity curves for the open and fretted notes, we see that the inharmonicity increases as we go up the fingerboard. Indeed, assuming that all other parameters remain constant, only the length $L$ _decreases_, resulting in a _larger_ $B$ and therefore larger deviations from the theoretical harmonics.


- __Playing the same note on different strings__: When playing the same note on two adjacent strings (e.g open vs. 5th fret) the inharmonicity/overall frequency deviation varies by a certain factor. The effect is even more pronounced if the strings are "further apart".


## II - INHARMONICITY CURVES FOR REAL RECORDINGS


We now compare the theoretical curves with the ones obtained with our recordings. For now, we do not take into account the effect of the guitar body. The idea is to __track__ through time the partials and their deviation from the ideal harmonics. Our approach is as follows: We first divide our recordings into successive __time frames__. Then, in each of these frames, we estimate the __partials' frequencies__ using one of the methods described below. Finally, we compute an approximate __inharmonicity coefficient__ $\hat{B}$ using all the estimated partials in all the frames.

### II.0 PARAMETERS


```python
# recording parameters
PATH_RECORDINGS = '../recordings/string_5/string_5_fret_0/'
file_name = 'string_5_fret_0_take_1.wav'

# read the recording
fs_recording, current_recording = wavfile.read(PATH_RECORDINGS + file_name)
y = current_recording / max(current_recording)

# we consider the first 1.5 seconds of the recording
y = y[:int(1.5*fs_recording)]

# time frames' parameters
tot_num_samp = y.shape[0]
frame_duration = .5
num_samp_per_frame = int(fs_recording * frame_duration)
window = np.hamming(num_samp_per_frame)
hop_size = int(num_samp_per_frame / 2)
num_frames = int((tot_num_samp - num_samp_per_frame) / hop_size)
```

### II.I TRACKING THE PARTIALS


In order to track the partials through successive time frames, we proceed as follows.

- We first estimate the fundamental frequency/pitch $f_0$ of the note that is being played (we assume monophonic sounds) using the whole recording. To that purpose, we use our pitch estimation method based on frequency ratios.


- The recording is then divided into overlapping, windowed frames (STFT-style). The amount of overlap, the type of window, and the duration of each frame are parameters. 


- Then, for each time frame, we estimate the spectrum in some pre-defined frequency regions by _zooming in_ around the ideal harmonics of the estimated fundamental. We have essentially three alternatives here, 

    - __regular DFT__: Perform a full DFT on each frame with a reasonable resolution. The problem is that we need a very fine frequency grid to be able to track the partials with precision and estimate the inharmonicity correctly, which requires a lot of computation time.
    
    - __modified regular MUSIC__: Use MUSIC to evaluate a frequency estimate of the spectrum in selected regions with a fine resolution. The advantage is that MUSIC makes some assumptions on the underlying signal instead of simply using the DFT. However, it is sensible to the parameters we use and not that efficient computationally.
    
    - __modified Goertzel's algorithm__: Evaluate the DFT in selected intervals with a fine resolution. The advantage is that we do not have to evaluate the whole DFT, thus reducing the computational burden. However, the frequency resolution is still limited by the window size. 


Mathematically, the estimated partials are obtained as local peaks, i.e

$$\hat{f}_k^{(i)} = arg \ max_{ \ f \in [k f_0 - \Delta_f, \ k f_0 + \Delta_f]} \ \Big({| P^{(i)}\Big(e^{j 2 \pi f}\Big) |^2 \Big)},$$

In the above, $i$ is the frame number, $k$ is the partial number, $\Delta_f$ is a small frequency interval (parameter) and $P^{(i)}\Big(e^{j 2 \pi f}\Big)$ is either the output of FFT, MUSIC or Goertzel's algorithm, depending on the method we used in the previous step.

### II.II INHARMONICITY ESTIMATION

From the estimated fundamental $f_0$ and the extracted partials $\hat{f}_k^{(i)}$, we aim to estimate the inharmonicity coefficient $\hat{B}$. Theoretically, from the expression

\begin{equation}
    f_n = n \ f_0 \ \sqrt{1 + B \ n^2}, 
\end{equation}

we can express $B$ as the coefficient of a degree-4 polynomial, as

\begin{equation}
    \Big(\frac{f_n}{f_0}\Big)^2 = n^2 + B \ n^4
\end{equation}

The above expression should hold for any partial number $n$. However, as we observed for the frequency difference, there will be some variation in practice. Thus, one approach is to use all the available estimated partials and perform a __least squares fit__ (a modified polynomial fit, essentially forcing some coefficients to 0).

Recall that the first $K$ partials have been estimated in $I$ frames. Thus, we have at our disposal a set $\{ \hat{f}_k^{(i)} \}_{i = 1, ..., I, k = 1, ..., K}$, where $i$ represents the frame index and $k$ the partial index. We can then form

\begin{equation}
\begin{bmatrix}\Big(\frac{\hat{f}_1^{(1)}}{\hat{f}_1^{(1)}}\Big)^2\\ ...\\ \Big(\frac{\hat{f}_K^{(1)}}{\hat{f}_1^{(1)}}\Big)^2\\ ...\\ \Big(\frac{\hat{f}_1^{(I)}}{\hat{f}_1^{(I)}}\Big)^2\\ ... \\ \Big(\frac{\hat{f}_K^{(I)}}{\hat{f}_1^{(I)}}\Big)^2\end{bmatrix} = \begin{bmatrix}1 & 1\\4 & 16\\ ... & ...\\ ... & ...\\ ... & ...\\ ... & ...\\ ... & ...\\ ... & ...\\ M^2 & M^4\end{bmatrix} \cdot \begin{bmatrix}\alpha\\\hat{B}\\\end{bmatrix}
\end{equation}

Solving the above system in a least squares fashion yields an estimate of the inharmonicity coefficient $\hat{B}$. 

__NB:__ Even though the theoretical motivations behind this 2D-LS are somewhat obscure, this method showed mure more coherent results than the 1D-LS. For example, we observe a monotonically increasing inharmonicity when moving up the fretboard, which agrees with the formula. It also gives less negative values. We thus decided to use this method. 

```python
from scipy.optimize import nnls

# Estimate the inharmonicity coefficient

# default method: 2D Least squares
partials_concat = [(estimated_partials_music[:, n]/estimated_partials_music[0, n])**2 for n in range(num_frames)]
LHS = np.hstack(partials_concat)
system_matrix = np.matlib.repmat(np.vstack((np.arange(1, max_mode + 1)**2, \
                                            np.arange(1, max_mode + 1)**4)).T, num_frames, 1)
coef, _, _, _ = np.linalg.lstsq(system_matrix, LHS)
estimated_inharm_coef = coef[1]  
print("\n\tEstimated inharmonicity coefficient: {b_hat}".format(b_hat=estimated_inharm_coef))

# alternative 1: 2D non-negative least squares (nnls)
coef_nnls, _ = nnls(system_matrix, LHS)
print("\n\t\tAlternative (NNLS): {b_hat}".format(b_hat=coef_nnls[1]))

# alternative 2: 1D Least squares / average
LHS = [((estimated_partials_goertzel[1:, n]/estimated_partials_goertzel[0, n])**2 - modes_num[1:]**2)/modes_num[1:]**4 for n in range(num_frames)]
print("\t\tAlternative 2 (1D-LS / Average): {b_hat}\n".format(b_hat=np.mean(np.hstack(LHS))))
```



Now, we would like a measure that is only string-dependent, i.e. that "identifies" the _string_ being played rather than the particular position used. Recall the expression of the inharmonicity coefficient $B$,
\begin{equation}
    B = \frac{\pi^2 E S K^2}{T L^2}
\end{equation}

Using this expression, we can simply multiply it by $L^2$ to keep only string-dependent properties ($E$, $S$, $K$ and $T$). We will consider a new constant $\gamma$ defined as

\begin{equation}
    \gamma = BL^2 = \frac{\pi^2 E S K^2}{T}
\end{equation}
